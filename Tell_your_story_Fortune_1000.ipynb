```python
#@title Default title text
import pandas as pd
import plotly.express as px
```

```python
url = 'https://drive.google.com/uc?id=1nG0CjxRkItiK-LqK3k8j1YOoRGaHuEUw'
df_full = pd.read_csv(url)
```

```python
df_full.head(3)
```

```python
df = df_full[['company', 'revenue', 'profit', 'sector', 'Market Cap']]

#let's rename num employees and Market cap 
df = df.rename({"Market Cap":"marketcap"}, axis = 1)

df.head()
```

```python
#checking how long our dataset is/how many companines we got 

print(f" We have {len(df)} companies in our dataset")
```

```python
#create a few more helper colums to assist us with colors and labels
#colors
df['color'] = df['sector'].apply(lambda x: "#CF5C36" if x in ['Financials', 'Technology'] else "#F4E3B2")

#labels
df["is_top_industry"] = df["sector"].apply(lambda x: "Yes" if x in ["Financials", "Technology"] else "No")

df.head(5)
```

```python
#Now let's make the scatter plot 
fig = px.scatter(df, 
                 x = "revenue",
                 y = "profit",
                 width = 1200,
                 height = 600,
                 size = "marketcap",
                 color = "is_top_industry",
                 category_orders = {"is_top_industry" : ["Yes", "No"]},
                 hover_data = ["company"],
                 labels = {
                    "revenue": "Company Revenue",
                    "profit": "Company Profit",
                    "is_top_industry":"Is it Technology or Finance",

                 },
                 title = "Technology and Financial companies are the Fortune 1K most profitable industries"
              )
fig.add_shape(type = "rect",
              x0=100000, y0=19600, x1=300000, y1=90000,
              line=dict(color="RoyalBlue")
              
              )
fig.add_shape(type = "line", x0=300000, x1=400000, y0=60000, y1=50000,line_color="RoyalBlue")

fig.add_annotation(
           x=450000,
           y=50000,
           text="Top 10 most profitable <br> companies are <br> in Technology or Financial",
           showarrow=False,
           font_size=15


)
              
          
                 
                 
                 
fig.show()
```
